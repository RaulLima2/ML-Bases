# Regressão Lógistica

A regressão logística é uma técnica de ML que temcomo objetivoproduzir, a partir de um conjunto de observações, ummodeloquepermita a predição de valores tomados por uma variável categórica, frequentemente binária, em função de uma ou mais variáveisindependentes contínuas e/ou binárias (GONZALES, 2018).

### Importação dos pacores

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

### Criação dos dados constantes

```python
n:int = 100
temp:list[float] = [random.uniform(27, 33) for _ in range(n)]
```

### Criação da função de regressão logistica

```python
def regression_logistic(alpha, beta)->np.ndarray:
    y:list = [np.random.binomial(1, 1 / (1 + np.exp(-alpha - beta * x))) for x in temp]
    plt.scatter(temp, y, marker='o')
    plt.xlabel('Temperatura')
    plt.ylabel('Resposta')
    x_vals: np.ndarray = np.linspace(min(temp), max(temp), 100)
    y_vals: np.ndarray = 1 / (1 + np.exp(-alpha - beta * x_vals))
    plt.plot(x_vals, y_vals, linewidth=2)
    plt.show()
    return y
```

### Teste

```python
y = regression_logistic(-88.585533, 2.972168)
```
